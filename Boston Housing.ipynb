```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()
for key, value in df.items():
    pi=sns.distplot(value, ax=ax[index])
    pi.set_xlabel(key, fontsize = 20)
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
from sklearn import preprocessing
cols = ["crim","zn","tax","black"] #since these values ae ot normalised
scalar= preprocessing.StandardScaler()
scaled_cols = scalar.fit_transform(df[cols])
scaled_cols = pd.DataFrame(scaled_cols, columns=cols)
for col in cols:
    df[col] = scaled_cols[col]
```

```python
fig, ax = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
ax = ax.flatten()
for key, value in df.items():
    pi=sns.distplot(value, ax=ax[index])
    pi.set_xlabel(key, fontsize = 20)
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
plt.figure(figsize=(20,10))
correlation = df.corr()
sns.heatmap(correlation,annot=True,cmap="coolwarm")
plt.show()
```

```python
x = df.drop(columns=["medv","lstat"],axis=1)
y = df["medv"]
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_squared_error
def train(model,x,y):
    x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42)
    model.fit(x_train,y_train)
    y_predict = model.predict(x_test)
    cv = cross_val_score(model,x,y,scoring="neg_mean_squared_error",cv=5)
    cv_score = np.abs(np.mean(cv))
    print("Report")
    print("Mean squared error:",mean_squared_error(y_predict,y_test))
    print("cv score:",cv_score)
```

```python
import xgboost as xgb
model = xgb.XGBRegressor()
train(model,x,y)
```
