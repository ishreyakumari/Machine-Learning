Using sklearn linear regression for Botson dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
boston = datasets.load_boston()
x = boston.data
y = boston.target
x = pd.DataFrame(x)
y = pd.DataFrame(y)
x.columns = boston.feature_names
xTrain,xTest,yTrain,yTest = model_selection.train_test_split(x,y)
linearRegression = LinearRegression()
linearRegression.fit(xTrain,yTrain)
yPred = linearRegression.predict(xTest)
plt.scatter(yPred,yTest)
plt.xlabel("Y Prediction")
plt.ylabel("Y Test")
m=[0,10,20,30,40,50]
plt.plot(m,m,color="black")
plt.show()
```
